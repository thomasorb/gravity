## Objectif

La sonde arrive dans l'atmosphère martienne avec une vitesse horizontale de 1000 m/s à une altitude 70 km.

Vous devez déterminer l'aire et le moment auquel ouvrir le parachute pour que l'atterrissage de la sonde sur Mars soit un succès.

Pour cela, vous avez plusieurs contraintes :

1 - la sonde doit atteindre la cible (symbolisée par une marque rouge au sol)

2 - la vitesse de la sonde au moment de l'impact ne doit pas dépasser 5 m/s

3 - la tension au niveau de l'accroche des parachutes ne doit pas dépasser 10000 N

4 - l'aire du parachute ne doit pas dépasser 1000 m²


```python
reload(simlib)

##### INPUT #######
T_DEPLOY  = 75 # instant auquel le parachute est déployé (en s)
AREA = 650 # aire du parachute
###################

MAX_STEPS = 30000
mars = simlib.Mars

p0 = (0, 70e3 + mars.radius) # (x0 (m), y0 (degrees))
v0 = (1000, 0) # 

probe = simlib.SpaceCraft(300, p0, v0, (0,0), 1, gridsize_min=0.2)

h, theta = mars.compute_htheta(probe.p)

step_nb = 0
while h > 0 or probe.t < 1:
    step_nb += 1
    #print(probe.t, probe.v)

    if probe.t > T_DEPLOY:
        probe.deploy_parachute(AREA)

    if h < 1:
        probe.detach_parachute()
    
    P, R, g, h = mars.get_PRgh(probe.p, probe.v)
    
    probe.update(P, R, g)

    if step_nb > MAX_STEPS:
       print('==== TOO MUCH STEPS ====')
       break

    # check landing velocity
    v = np.sqrt(np.sum(probe.v**2))
    if h <= 0:
        if v > 5:
            print('==== PROBE CRASHED !!! ====')
            print('Final velocity: {:.2f} m/s'.format(v))
        else:
            print('LANDING SUCCESSFUL')
            print('Final velocity: {:.2f} m/s'.format(v))
        

fig = plt.figure()

probe.plot_trajectory(mars, theta_target=(1.8,1.9))

probe.plot_parachute_tension(max_tension=10e3)

probe.plot_velocity()

probe.plot_acceleration()
```
